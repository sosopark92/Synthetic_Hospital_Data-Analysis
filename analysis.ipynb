## Clean Data
This step focuses on preparing raw Synthea EHR data for analysis by standardising schemas and data types while preserving clinically
meaningful missing values.
- correct data types
- a small set of analytics-ready tables
- clinically meaningful missing values preserved

```python
patients.columns
```

```python
encounters.columns
```

```python
conditions.columns
```

```python
patients_clean = patients[["Id","BIRTHDATE", "GENDER"]].copy()

patients_clean.head()
```

```python
encounters_clean = encounters[
    [
        "Id",
        "PATIENT",
        "START",
        "STOP",
        "ENCOUNTERCLASS",
        "DESCRIPTION",
        "BASE_ENCOUNTER_COST",
        "TOTAL_CLAIM_COST"
    ]
].copy()

encounters_clean.head()
```

```python
conditions_clean = conditions[
    ["PATIENT", "ENCOUNTER", "DESCRIPTION"]
].copy()

conditions_clean.head()
```

```python
# Convert datetime data to correct datatype
patients_clean["BIRTHDATE"] = pd.to_datetime(patients_clean["BIRTHDATE"], errors="coerce")
encounters_clean["START"] = pd.to_datetime(encounters_clean["START"], errors="coerce")
encounters_clean["STOP"] = pd.to_datetime(encounters_clean["STOP"], errors="coerce")
```

```python
# Length of stay
encounters_clean["LENGTH_OF_STAY_DAYS"] = (encounters_clean["STOP"] - encounters_clean["START"]).dt.days
```

```python
encounters_clean.to_csv(
    "analytics_table.csv",
    index=False
)
```

## General Analysis & Visualisation

### Length of Stay Distribution

```python
encounters_clean["LENGTH_OF_STAY_DAYS"].describe()
```

```python
encounters_clean[encounters_clean["LENGTH_OF_STAY_DAYS"] > 0]["LENGTH_OF_STAY_DAYS"].plot(
    kind="hist", bins=30, title="Distribution of Inpatient Length of Stay"
)
```

✅ Most encounters have a length of stay of 0 days.\
✅ A small number of encounters have longer stays (up to 61 days)

### Encounter Cases

```python
encounters_clean["ENCOUNTERCLASS"].value_counts().plot(
    kind="bar", title="Encounter Type Distribution"
)
```

✅ Most encounters are ambulatory and wellness visits.

```python
encounters_clean["YEAR"] = encounters_clean["START"].dt.year
encounters_clean["MONTH"] = encounters_clean["START"].dt.month
```

```python
yearly_demand = encounters_clean.groupby("YEAR").size()

yearly_demand.plot(
    kind="line",
    marker="o",
    title="Annual Hospital Encounter Volume"
)
```

✅ Overall demand shows a clear increase, with some year-to-year variation, suggesting growing pressure on hospital services.

```python
yearly_demand[yearly_demand.index >= 2015].plot(
    kind="line",
    marker="o",
    title="Annual Hospital Encounter Volume"
)
```

✅ Demand increases over time, with a noticeable peak around 2021 followed by a stabilisation at a higher level.

```python
max_year = encounters_clean["YEAR"].max()
max_year
```

```python
latest_year = encounters_clean[encounters_clean["YEAR"] == max_year]
latest_year.shape
```

```python
latest_year["MONTH"] = latest_year["START"].dt.month
monthly = latest_year.groupby("MONTH").size()
```

```python
monthly.plot(
    kind="bar",
    title=f"Monthly Hospital Encounter Volume ({max_year})"
)
```

✅ Demand rises through the middle of the year and declines toward the end.

```python
recent_3y = encounters_clean[
    encounters_clean["YEAR"] >= (max_year - 2)
]

recent_3y["YEAR"].unique()
```

```python
monthly_3y = (
    recent_3y
    .groupby(["YEAR", "MONTH"])
    .size()
    .reset_index(name="ENCOUNTERS")
)

monthly_3y.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()

for y in sorted(monthly_3y["YEAR"].unique()):
    data = monthly_3y[monthly_3y["YEAR"] == y]
    plt.plot(
        data["MONTH"],
        data["ENCOUNTERS"],
        marker="o",
        label=str(y)
    )

plt.title("Monthly Hospital Encounter Volume (Recent 3 Years)")
plt.xlabel("Month")
plt.ylabel("Number of Encounters")
plt.xticks(range(1, 13))
plt.legend(title="Year")
plt.show()
```

✅ Although monthly variation is observed, the pattern is not consistent across the most recent three years. This suggests that hospital demand does not exhibit a strong or predictable seasonal trend.

### Top 10 Diagnoses Driving Longer Stays

```python
los_by_diag = (
    encounters_clean
    .dropna(subset=["DESCRIPTION"])
    .groupby("DESCRIPTION")["LENGTH_OF_STAY_DAYS"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

los_by_diag.plot(kind="barh", title="Diagnoses with Longest Average Stay")
```

✅ Admissions related to hospice care, isolation, rehabilitation, and intensive or specialist care show substantially longer stays, indicating higher resource intensity despite lower encounter volumes.

# Healthcare Utilisation & Cost Analysis

### Total Claim Cost by Encounter Type

```python
import matplotlib.ticker as ticker

ax = encounters_clean.groupby("ENCOUNTERCLASS")["TOTAL_CLAIM_COST"].sum().plot(
    kind="bar", title="Total Claim Cost by Encounter Type"
)

ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f"${x/1_000_000}M")
)

ax.set_ylabel("Total Claim Cost (USD)")
```

```python
import matplotlib.ticker as ticker

ax = encounters_clean.groupby("ENCOUNTERCLASS")["TOTAL_CLAIM_COST"].mean().plot(
    kind="bar", title="Average Claim Cost by Encounter Type"
)

ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f"${x/1_000}K")
)

ax.set_ylabel("Average Claim Cost (USD)")
```

✅ Ambulatory encounters dominate total costs but not average costs. Their average claim cost per encounter is relatively low compared to inpatient, SNF, or hospice care. This indicates that ambulatory services are high-frequency, low-cost encounters.

✅ Inpatient encounters show the highest average claim cost across all encounter types. Despite this, their total claim cost is significantly lower than ambulatory encounters. Inpatient care is low-frequency but high-cost.

✅ Skilled Nursing Facility (SNF) and hospice encounters have high average claim costs, second only to inpatient care. Their total costs remain moderate due to lower encounter volumes.

## Encounter Volume vs Cost

```python
encounter_cost = (
    encounters_clean
    .groupby("ENCOUNTERCLASS")["TOTAL_CLAIM_COST"]
    .agg(
        encounter_volume="count",
        total_cost="sum"
    )
    .reset_index()
)

encounter_cost
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

#creates a scatter plot
ax.scatter(
    encounter_cost["encounter_volume"],
    encounter_cost["total_cost"]
)

# Label each point with encounter type
for _, row in encounter_cost.iterrows():
    ax.text(
        row["encounter_volume"],
        row["total_cost"],
        row["ENCOUNTERCLASS"],
        fontsize=9,
        ha="left",
        va="bottom"
    )

# Axis formatting
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, d: f"${x/1_000_000:.1f}M")
)

ax.set_xlabel("Encounter Volume (Number of Visits)")
ax.set_ylabel("Total Claim Cost (USD)")
ax.set_title("Encounter Volume vs Total Claim Cost")

plt.tight_layout()
plt.show()
```

## Top 10% High-Cost Patients Contribution

```python
import pandas as pd

# Total cost per patient
patient_cost = (
    encounters_clean
    .groupby("PATIENT")["TOTAL_CLAIM_COST"]
    .sum()
    .reset_index()
)

# Sort patients by cost
patient_cost = patient_cost.sort_values("TOTAL_CLAIM_COST", ascending=False)

# Identify top 10%
top_10 = int(len(patient_cost) * 0.10)

top_10_patients = patient_cost.iloc[:top_10]
remaining_patients = patient_cost.iloc[top_10:]

# Cost contribution
top_10_cost = top_10_patients["TOTAL_CLAIM_COST"].sum()
remaining_cost = remaining_patients["TOTAL_CLAIM_COST"].sum()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

labels = ["Top 10% Patients", "Remaining 90% Patients"]
values = [top_10_cost, remaining_cost]

fig, ax = plt.subplots(figsize=(6, 4))

ax.bar(labels, values)

ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, pos: f"${x/1_000_000:.1f}M")
)

ax.set_title("Healthcare Cost Contribution by Patient Group")
ax.set_ylabel("Total Claim Cost (USD)")

plt.tight_layout()
plt.show()
```

✅ The top 10% of patients account for a large share of total healthcare costs.This indicates a highly skewed cost distribution, where healthcare spending is concentrated among a small number of high-cost patients.

# Join the tables and Export CSV for Power BI

```python
patients_clean.head(1)
```

```python
encounters_clean.head(1)
```

```python
# Match each encounter’s patient ID with the patient ID in the patient table.

analytics = encounters_clean.merge(
    patients_clean,
    left_on="PATIENT",
    right_on="Id",
    how="left",
    suffixes=("", "_PATIENT")
)

analytics.shape
```

✅ To integrate patient and encounter data, a left join was applied using the patient identifier as the key.

```python
analytics.head(1)
```

```python
conditions_clean.head(1)
```

```python
analytics = analytics.merge(
    conditions_clean,
    left_on="Id",
    right_on="ENCOUNTER",
    how="left",
    suffixes=("", "_CONDITION")
)

analytics.shape
```

```python
# Add time features
analytics["START"] = pd.to_datetime(analytics["START"], errors="coerce")
analytics["YEAR"] = analytics["START"].dt.year
analytics["MONTH"] = analytics["START"].dt.month
```
